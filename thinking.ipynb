```python
import ollama

# Configure your server URL here
SERVER_HOST = 'http://ollama.cs.wallawalla.edu:11434'
client = ollama.Client(host=SERVER_HOST)

def gwr(task, model='deepseek-r1'):
    """Generate code with reasoning."""
    client = ollama.Client(host='http://ollama.cs.wallawalla.edu:11434')
    
    response = client.chat(
        model=model,
        messages=[{'role': 'user', 'content': task}],
        think=True
    )
    
    message = response['message']
    thinking, output =  message.get('thinking', ''), message.get('content', '')
    print("OUTPUT\n" + "="*60)
    print(f"TASK: ")
    print("-" * 60)
    print(task)
    print()
    
    print("REASONING:")
    print("-" * 60)

    print(thinking)
    
    print(f"\nFINAL OUTPUT:")
    print("-" * 60)

    print(output)
    print("="*60)
```

```python
%time gwr("Solve 2x + 5 = 17")
```

```python
%time gwr("A train travels 120 miles in 2 hours. What's its speed?")
```

```python
%time gwr("A farmer has 17 sheep. All but 9 run away. How many sheep are left?")
```
